**Import required libraries**

```python
import pandas as pd
```

**Load the dataset**

```python
df = pd.read_csv("Sample - Superstore.csv", encoding="latin1")
df.head()
```

**Total Sales by Region**

```python
df.groupby("Region")["Sales"].sum().reset_index()
```

**Average Profit by Category**

```python
df.groupby("Category")["Profit"].mean().reset_index()
```

**Order Count by Ship Mode**

```python
df.groupby("Ship Mode")["Order ID"].count().reset_index()
```

**Region-wise Sales & Profit**

```python
df.groupby("Region").agg({"Sales": "sum", "Profit": "sum"}).reset_index()
```
